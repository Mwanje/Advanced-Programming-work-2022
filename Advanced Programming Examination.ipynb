k) Consider the variable B ={“WHO Region”, “Geographical location”, “longitude and
latitude” }. Write python code that can change WHO Region to “Region”, Geographical
location to “Location”, longitude and latitude to “Location”. (4 marks)

```python
# this is my sample data frame
df = pd.DataFrame({'WHO Region': ["Central", "Eastern", "Western"],
                    
                   'Geographical location': ["Kampala", "Jinja", "Mbarara"],
                    
                   'longitude and latitude': [2.774663568, 32.2999988, 1.3733]})
 
# print data frame
display(df)
  
dict = {'WHO Region': 'Region',
        'Geographical location': 'Location',
        'longitude and latitude': 'Location'}
 
# call rename () method
df.rename(columns=dict,
          inplace=True)
 
# print Output Data frame after rename columns
display(df)
```
