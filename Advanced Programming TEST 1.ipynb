```python
#defining imports 
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data from url
data = requests.get('https://www.worldometers.info/world-population/uganda-population/')
data_tables = pd.read_html(data.text)
```

```python
#lets see the tables from this data
data_tables
```

```python
#reading the data set for Population of Uganda
table_df = data_tables[1]
table_df
```

```python
#removing % sign
cols = ['Yearly % Change','Urban Pop %',"Country's Share of World Pop"]
table_df[cols] = table_df[cols].replace({'%': ''}, regex=True)
```

# (a). Compute the maximum value of each feature

```python
max_values = table_df.max()
print('The maximum values of every variable are:\n', max_values)
```

# (b). Compute the minimum value of each feature

```python
min_values = table_df.min()
print('The minimum values of every variable are:\n', min_values)
```

# (c). Compute the mean value of each feature 

```python
mean_values = table_df.mean(numeric_only=True)
print('The mean values of every variable are:\n', mean_values)
```

# (d). Compute the standard deviation of each feature 

```python
std_values = table_df.std(numeric_only=True)
print('The standard deviation of every variable are:\n', std_values)
```

# (e). Evaluate the correlations between features

```python
table_df.corr(method ='pearson')
```

```python
#this figure helps us to relate the correelation between features
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(table_df.corr(method ='pearson'), annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```
